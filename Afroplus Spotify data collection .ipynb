```python
tracks_data = []

for _, row in artists_df.iterrows():
    try:
        # Get top tracks for each artist
        top_tracks = sp.artist_top_tracks(row["spotify_id"], country="CM")
        
        for track in top_tracks["tracks"]:
            data = {
                "artist_name": row["artist_name"],
                "track_name": track["name"],
                "popularity": track["popularity"],
                "album": track["album"]["name"],
                "release_date": track["album"]["release_date"],
                "duration_ms": track["duration_ms"],
                "explicit": track["explicit"],
                "track_id": track["id"]
            }
            tracks_data.append(data)
        
        print(f"✓ {row['artist_name']} — {len(top_tracks['tracks'])} tracks found")
        time.sleep(0.5)

    except Exception as e:
        print(f"✗ Failed for {row['artist_name']}: {e}")

tracks_df = pd.DataFrame(tracks_data)
print(f"\nTotal tracks collected: {len(tracks_df)}")
tracks_df.head()
```
